```python
## Independent and Dependent Features

X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Standardize the dataset

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

# Model Training

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
# Print the cofficient and intercept

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameters the model has been train
regression.get_params()
```

```python
# Prediction with Test data

reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

## Assumptions

```python
# ploat a scatter ploat for the Prediction
plt.scatter(y_test,reg_predict)
```

```python
# Residulas (Error)
residuals = y_test - reg_predict
residuals
```

```python
# ploat the Residuals

sns.displot(residuals,kind="kde")
```

```python
# Scatter ploat with respect to prediction an residuals
# uniform disribution
plt.scatter(reg_predict,residuals)
```

```python
# finding the Error

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print("Mean Squared Error:",mean_squared_error(y_test,reg_predict))
print("Mean Absolute Error:",mean_absolute_error(y_test,reg_predict))
print("Root Mean Squared Error:",np.sqrt(mean_squared_error(y_test,reg_predict)))
```

## R square and adjusted R square

Formula <br>
**R^2 = 1 - SSR/SST** <br> <br>
R^2 = Coefficient of determinstion <br> 
SSR = sum of squares of residuals <br> 
SST = total sum of squares

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,reg_predict)
print("R2 Score:",score)
```

**Adjusted R2 = 1  [(1-R2)*(n-1)/(n-k-1)]**

where:

R2: The R2 of the model <br>
n: The number of observations <br>
k: The number of predictor variables

```python
# display adjusted R-squared

1- (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1] -1)
```
